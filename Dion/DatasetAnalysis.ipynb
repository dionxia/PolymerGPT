```python
import pandas as pd

# File paths of the datasets you want to analyze
file_paths = {
    #'PE_I': '/home/dion/projects/New_TransPolymer/TransPolymer/data/PE_I.csv',
    #'PE_II': '/home/dion/projects/New_TransPolymer/TransPolymer/data/PE_II.csv',
    'Egc': '/home/dion/projects/New_TransPolymer/TransPolymer/data/Egc.csv',
    'Egb': '/home/dion/projects/New_TransPolymer/TransPolymer/data/Egb.csv',
    'Eea': '/home/dion/projects/New_TransPolymer/TransPolymer/data/Eea.csv',
    'Ei': '/home/dion/projects/New_TransPolymer/TransPolymer/data/Ei.csv',
    'Xc': '/home/dion/projects/New_TransPolymer/TransPolymer/data/Xc.csv',
    'EPS': '/home/dion/projects/New_TransPolymer/TransPolymer/data/EPS.csv',
    'Nc': '/home/dion/projects/New_TransPolymer/TransPolymer/data/Nc.csv',
    #'OPV': '/home/dion/projects/New_TransPolymer/TransPolymer/data/OPV.csv'
}

# Load datasets into a dictionary
datasets = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Assuming the column with polymer strings is named 'polymer_string' in all datasets
# If the column name is different in some datasets, you can adjust accordingly

# Add a 'dataset' column to each dataset to track its origin
for name, df in datasets.items():
    df['dataset'] = name

# Concatenate all datasets into a single DataFrame
all_data = pd.concat(datasets.values(), ignore_index=True)

# Group by the polymer strings and count unique dataset appearances
polymer_counts = all_data.groupby('smiles')['dataset'].nunique().reset_index()

# Filter polymers that appear in more than one dataset
duplicate_polymers = polymer_counts[polymer_counts['dataset'] > 1]

# Group by polymer strings and aggregate the datasets where each polymer appears
polymer_datasets = all_data.groupby('smiles')['dataset'].agg(list).reset_index()

# Join this information with the duplicate polymer counts (optional, but useful for seeing how many datasets)
polymer_datasets['dataset_count'] = polymer_datasets['dataset'].apply(len)

# Filter for polymers that appear in more than one dataset
duplicate_polymers_with_datasets = polymer_datasets[polymer_datasets['dataset_count'] > 1]

# Display the result
display(duplicate_polymers_with_datasets)
```

```python
# Group by polymer strings and aggregate the datasets where each polymer appears
polymer_datasets = all_data.groupby('smiles')['dataset'].agg(list).reset_index()

# Add a column for the number of datasets (i.e., how many datasets each polymer appears in)
polymer_datasets['dataset_count'] = polymer_datasets['dataset'].apply(len)

# Filter for polymers that appear in more than one dataset (optional, depending on whether you want only duplicates)
duplicate_polymers_with_datasets = polymer_datasets[polymer_datasets['dataset_count'] > 1]

# Count how many polymers appear in each number of datasets (4, 5, 6, etc.)
dataset_count_summary = duplicate_polymers_with_datasets['dataset_count'].value_counts().sort_index()

# Display the summary
print(dataset_count_summary)
```
